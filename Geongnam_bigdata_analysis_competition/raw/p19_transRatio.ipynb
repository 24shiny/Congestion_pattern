```python
import pandas as pd
df = pd.read_csv('평일전국교통량행렬.csv',encoding='utf-8')
df.head()
```

```python
df = df.rename(columns={'Unnamed: 0':'name'})
df = df.set_index('name')
df.head()
```

```python
# clear self-loops
for i in range(len(df.iloc[0])):
    df.iloc[i][i] = 0
```

```python
df
```

```python
# sum columns up
totalTraff = df.sum(axis='columns')
totalTraff
```

```python
totalTraffRow = df.sum(axis='rows')
totalTraffRow
```

```python
df.loc[-1] = totalTraffRow.values
df
```

```python
df = df.rename(index={-1:"entTotalTraffic"})
```

```python
df['exitTotalTraff'] = totalTraff
df
```

```python
# first check point
df.to_csv('adjMatEntExit.csv',encoding='utf-8')
```

```python
# copy the df and replace values by their ratio to total traffic
# ratio dataFrame
rdf = df.copy()
```

```python
rdf['경기도'] = round(rdf['경기도']/rdf['exitTotalTraff'],3)
rdf['경상남도'] = round(rdf['경상남도']/rdf['exitTotalTraff'],3)
rdf['충청남도'] = round(rdf['충청남도']/rdf['exitTotalTraff'],3)
rdf['충청북도'] = round(rdf['충청북도']/rdf['exitTotalTraff'],3)
rdf['대전광역시'] = round(rdf['대전광역시']/rdf['exitTotalTraff'],3)
rdf['경상북도'] = round(rdf['경상북도']/rdf['exitTotalTraff'],3)
rdf['대구광역시'] = round(rdf['대구광역시']/rdf['exitTotalTraff'],3)
rdf['울산광역시'] = round(rdf['울산광역시']/rdf['exitTotalTraff'],3)
rdf['부산광역시'] = round(rdf['부산광역시']/rdf['exitTotalTraff'],3)
rdf['전라북도'] = round(rdf['전라북도']/rdf['exitTotalTraff'],3)
rdf['전라남도'] = round(rdf['전라남도']/rdf['exitTotalTraff'],3)
rdf['강원도'] = round(rdf['강원도']/rdf['exitTotalTraff'],3)
rdf['광주광역시'] = round(rdf['광주광역시']/rdf['exitTotalTraff'],3)
rdf['세종특별자치시'] = round(rdf['세종특별자치시']/rdf['exitTotalTraff'],3)
```

```python
rdf
```

```python
rdf.to_csv('exitRatioAdjMatrix.csv',encoding='utf-8')
```
